# Gradient Descent In Machine Learning

```python
m=0
c=0
learning_rate=0.001
n=int(len(x))

iterations=1000
for i in range(iterations):
    y_prediction=m*x+c
    derivates_m=-(2/n)*+(x*(y- y_prediction))
    derivates_c=-(2/n)*+((y- y_prediction))
    m= m - learning_rate* derivates_m
    c= c - learning_rate* derivates_c 
    
print(m,c)
```
